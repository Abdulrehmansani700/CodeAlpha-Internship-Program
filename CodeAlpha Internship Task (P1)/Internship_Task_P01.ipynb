```python
import random
```

```python
word = ["pyhton" , "hangman", "programming", "computer" , "laptop"]
```

```python
HANGMAN_STAGES = [
    """
  +---+
  |   |
      |
      |
      |
      |
=========""",
    """
  +---+
  |   |
  O   |
      |
      |
      |
=========""",
    """
  +---+
  |   |
  O   |
  |   |
      |
      |
=========""",
    """
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========""",
    """
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========""",
    """
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========""",
    """
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========""",
]

MAX_WRONG = 6
```

```python
def display_state(wrong_count, guessed_letters, secret_word):
    print(HANGMAN_STAGES[wrong_count])
    print(f"wrong_guesses_left: {MAX_WRONG -wrong_count}")
    print(f"letters_guessed: {', '.join(sorted(guessed_letters)) if guessed_letters else '-'}")
    display_word = " " .join(letter if letter in guessed_letters else "_" for letter in secret_word)
    print(display_word)
```

```python
def get_player_guess(guessed_letters):
    while True:
        guess = input("guess a letter: ").strip().lower()
        if len(guess) != 1:
            print("one letter")
        elif not guess.isalpha():
            print("one alphabet")
        elif guess in guessed_letters:
            print("the letter is already guess")
        else:
            return guess     
```

```python
def play_hangman():
    secret_word = random.choice(word)
    guessed_letters = set()
    wrong_count = 0

    print("=" * 40)
    print("hangman game - codealpha")
    print("=" * 40)
    print(f"word in {len(secret_word)} letters. good luck")

    while True:
        display_state(wrong_count,guessed_letters,secret_word)
        if all(letter in guessed_letters for letter in secret_word):
            print("congragulations:", secret_word.upper())
            break
        guess = get_player_guess(guessed_letters)
        guessed_letters.add(guess)

        if guess in secret_word:
            print("right guess")
        else:
            wrong_count += 1
            print("false '{guess}'is not in word.")
```

```python
def main():
    while True:
        play_hangman()
        again = input("agin playing(y/n):").strip().lower()
        if again not in ("y","y"):
            print("thanks")
            break
if __name__ == "__main__":
    main()
```
