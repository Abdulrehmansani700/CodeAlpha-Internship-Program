```python
STOCK_PRICES = {
    "AAPL":  180,
    "TSLA":  250,
    "GOOGL": 140,
    "AMZN":  175,
    "MSFT":  380,
}
```

```python
def show_available_stocks():
    print("=" * 40)
    print("   AVAILABLE STOCKS")
    print("=" * 40)
    
    for symbol, price in STOCK_PRICES.items():
        print(f"  {symbol}  →  ${price}")
    
    print("=" * 40)
```

```python
def get_portfolio():
    portfolio = {}

    print("Stock symbol daalo, khatam karne ke liye 'done' likho")

    while True:
        symbol = input("Stock symbol: ").strip().upper()

        if symbol == "DONE":
            break

        if symbol not in STOCK_PRICES:
            print("Yeh stock available nahi!")
            continue

        quantity = int(input("Quantity: "))
        portfolio[symbol] = quantity

    return portfolio
```

```python
def calculate_investment(portfolio):
    total = 0

    for symbol, quantity in portfolio.items():
        price = STOCK_PRICES[symbol]
        stock_total = price * quantity
        total += stock_total
        print(f"{symbol}  x{quantity}  ${price}  =  ${stock_total}")

    print(f"TOTAL = ${total}")
    return total
```

```python
def save_to_file(portfolio, total):
    save = input("File mein save karna hai? (yes/no): ")

    if save != "yes":
        print("Theek hai, save nahi kiya")
        return

    with open("portfolio_result.txt", "w") as f:
        for symbol, quantity in portfolio.items():
            price = STOCK_PRICES[symbol]
            stock_total = price * quantity
            f.write(f"{symbol} x{quantity} = ${stock_total}\n")
        
        f.write(f"TOTAL = ${total}\n")

    print("File save ho gayi!")
```

```python
def main():
    show_available_stocks()    # Step 1: stocks dikhao
    portfolio = get_portfolio() # Step 2: user se lo
    total = calculate_investment(portfolio)  # Step 3: calculate karo
    save_to_file(portfolio, total)  # Step 4: save karo

if __name__ == "__main__":
    main()
```
